## Subset CMIP6 Datasets with xarray from web 
web: https://esgf-pyclient.readthedocs.io/en/latest/notebooks/demo/subset-cmip6.html

```python
import numpy as np
import xarray as xr
```

```python
from pyesgf.search import SearchConnection
conn = SearchConnection('https://esgf-data.dkrz.de/esg-search', distrib=True)
```

```python
ctx = conn.new_context(
    project='CMIP6',
    source_id='UKESM1-0-LL',
    experiment_id='historical',
    variable='tas',
    frequency='mon',
    variant_label='r1i1p1f2',
    data_node='esgf-data3.ceda.ac.uk')
ctx.hit_count
```

```python
result = ctx.search()[0]
result.dataset_id
```

```python
files = result.file_context().search()
for file in files:
    print(file.opendap_url)
```

# Subset single dataset with xarray

```python
ds = xr.open_dataset(files[0].opendap_url)
ds
```

```python
da = ds['tas']
da = da.isel(time=slice(0, 1))
da = da.sel(lat=slice(-50, 50), lon=slice(0, 50))
```

```python
da.plot()
```

```python
%matplotlib inline
da.plot()
```

```python
%matplotlib inline
```

```python
da.to_netcdf('tas_africa_19500116.nc')

```
